```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple
```

```python
n_groups = 4
Ads_energy = [-4.88, -7.17, -3.82, -0.12]
y = [-4.88, -7.17, -3.82, -0.12]
fig, ax = plt.subplots(figsize=(4,3),dpi=200)
index = np.arange(n_groups)
bar_width = 0.35
#opacity = 0.4
#error_config = {'ecolor': '0.3'}
rects1 = ax.bar(index, Ads_energy, bar_width,
                color='r')
ax.set_ylabel(' Adsorbtion Energy E$_{\mathrm{ads}}$ (eV)', fontsize=10)
#ax.set_title('Scores by group and gender')
ax.set_xticks(index )
ax.set_xticklabels(('CoS', 'FeS', 'NiS', 'CuS'),rotation=0)
#ax.set_xlabel('Regression Methods', fontsize=12)
#ax.legend()
fig.tight_layout()
ax.invert_yaxis()
plt.margins(y=0.1)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
xlocs, xlabs = plt.xticks()
for i, v in enumerate(y):
    plt.text(xlocs[i] - 0.22, v - .09, str(v))
plt.savefig ('H2O2_Adsorption.eps')
plt.show()
# from google.colab import files
# files.download('H2O2_Adsorption.png') 
```

```python
n_groups = 4
Adb_OH = (-3.36, -3.23, -2.37, -0.40)
y1 = (-3.36, -3.23, -2.37, -0.40)
Adb_2OH = (-3.50, -5.68, -3.47, -0.37)
y2 = (-3.50, -5.68, -3.47, -0.37)
fig, ax = plt.subplots(figsize=(4,3),dpi=400)
index = np.arange(n_groups)
bar_width = 0.35
rects1 = ax.bar(index, Adb_OH, bar_width,
                color='b', label='MS-OH')
rects2 = ax.bar(index+ bar_width, Adb_2OH, bar_width,
                color='r', label='MS-2OH')
#ax.set_xlabel('Group')
ax.set_ylabel(' Adsorbtion Energy E$_{\mathrm{ ads}}$ (eV)', fontsize=10)
#ax.set_title('Scores by group and gender')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('CoS', 'FeS', 'NiS', 'CuS'),rotation=0)
#ax.set_xlabel('Regression Methods', fontsize=12)
ax.invert_yaxis()
plt.margins(y=0.1)
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.legend(loc=2, prop={'size': 13}, frameon=True)
xlocs, xlabs = plt.xticks()
for i, v in enumerate(y1):
    plt.text(xlocs[i] - 0.45, v - .09, str(v), fontsize=8)
for i, v in enumerate(y2):
    plt.text(xlocs[i] - 0.03, v - .09, str(v),fontsize=8)    
leg = plt.legend()
leg.get_frame().set_edgecolor('black')
fig.tight_layout()
plt.savefig ('OH-2OH.eps')
plt.show()
# from google.colab import files
# files.download('OH-2OH.eps') 
```
